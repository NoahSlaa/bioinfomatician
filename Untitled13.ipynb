```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# URL for the daily reports from Johns Hopkins CSSEGISandData GitHub repository
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-01-2023.csv' # Example date, you might want to dynamically get the latest date

try:
    df = pd.read_csv(url)
    print(f"Data loaded successfully from {url}")
except Exception as e:
    print(f"Error loading data: {e}")
    print("Please check the URL or try a different date format/source if the data is not available for this specific date.")

print('CLEANED DATA.................for east africa...............................................FOR VISUALIZATION')
east_africa=['Tanzania','Kenya','Uganda','Rwanda','Burundi','South Sudan']
data_east=df[df['Country_Region'].isin(east_africa)].reset_index(drop=True).fillna(0)
selected=['Country_Region','Deaths','Case_Fatality_Ratio','Confirmed','Last_Update']
filtred=data_east[selected]
x=filtred.sort_values(by='Country_Region',ascending=True).reset_index(drop=True)
display(x.head(5))


print('CLEANED ........................for west africa..............................................FOR VISUALIZATION')
west_africa=['Benin','Cape Vedre','Gambia','Ghana','Guinea','Guinea Bissau','Liberia','Mali','Niger','Nigeria','Senegal','Sierra Leone','Togo']
data_west=df[df['Country_Region'].isin(west_africa)].reset_index(drop=True).fillna(0)
clean=['Country_Region','Deaths','Case_Fatality_Ratio','Confirmed','Last_Update']
simple=data_west[clean]
sorted=simple.sort_values(by='Country_Region',ascending=True).reset_index(drop=True)
display(sorted.head(5))

print('ACTUAL CALCULATION PERTAINING TO DATA ')
x['Death_percentage']=np.divide(x['Deaths'],x['Confirmed'])*100
x['Survival-Percentage']=np.subtract(100,x['Death_percentage'])
x['Deaths per mortality rate']=np.divide(x['Deaths'],x['Case_Fatality_Ratio'])
display(x.head(5))

print('ACTUAL CALCULATION PERTAINING TO DATA ')
sorted['Death_percentage']=np.divide(sorted['Deaths'],sorted['Confirmed'])*100
sorted['Survival-Percentage']=np.subtract(100,sorted['Death_percentage'])
sorted['Death per mortality rate']=np.divide(sorted['Deaths'],sorted['Case_Fatality_Ratio'])
display(sorted.head(5))
print('DATA VISUALIZATION')


sns.set_theme(style='whitegrid')
fig,axes=plt.subplots(2,2,figsize=(15,10))
axes=axes.flatten()
print('FOR EAST AFRICA')
sns.barplot(x='Country_Region',y='Survival-Percentage',data=x,ax=axes[0],palette='viridis')
axes[0].set_title('DEATH PERCENTAGE PER CASES CONFIRMED IN COVID-19 OUTBREAK')
axes[0].set_xlabel('Countries')
axes[0].set_ylabel('Percentage(%)')

sns.barplot(x='Country_Region',y='Survival-Percentage',data=sorted,ax=axes[1],palette='magma')
axes[1].set_title('SURVIVAL PERCENTAGE PER CASES CONFIRMED IN COVID-19 OUTBREAK')
axes[1].set_xlabel('Countries')
axes[1].set_ylabel('Percentage(%)')
axes[1].tick_params(axis='x',rotation=90)

plt.tight_layout()
plt.show()







```

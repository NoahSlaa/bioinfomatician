```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_colwidth",None)
df=pd.read_csv("Gavin_PPI.txt",sep="\t",header=None)
data=df[0].head(100).value_counts().reset_index()
data.columns=["Gene","Frequency"]

#DATA VISUALIZATION
top=data.head(7)
sns.set_theme(style="white")
color=sns.color_palette("pastel")


plt.figure(figsize=(6,6),dpi=300)
plt.pie(
    top["Frequency"],
    labels=top["Gene"],
    colors=color,
    autopct="%0.1f%%",
    startangle=140,
    wedgeprops={"edgecolor":"white","linewidth":2,"antialiased":True},
    textprops={"fontsize":12,"fontweight":"bold"})

centre_circle=plt.Circle((0,0),0.50,color="white")
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("TOP SEVEN GENES AND THEIR CONTRIBUTION",weight="bold")


plt.show()
```
